# API some country info  _______________

```python
### imports
import requests

import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
### configuratories
gdp_endpoint = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json&per_page=20000'
life_endpoint = 'https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?format=json&per_page=20000'
country_info_endpoint = 'https://api.worldbank.org/v2/country?format=json&per_page=20000'
USA_country_info_endpoint = 'https://api.worldbank.org/v2/country/us?format=json'

### bonus
### 1: only usa
USA_gdp_endpoint = 'https://api.worldbank.org/v2/country/USA/indicator/NY.GDP.PCAP.PP.KD?format=json&per_page=20000'
USA_life_endpoint = 'https://api.worldbank.org/v2/country/USA/indicator/SP.DYN.LE00.IN?format=json&per_page=20000'
### 2: only usa 2021
USA2021_gdp_endpoint = 'https://api.worldbank.org/v2/country/USA/indicator/NY.GDP.PCAP.PP.KD?format=json&date=2021&per_page=20000'
USA2021_life_endpoint = 'https://api.worldbank.org/v2/country/USA/indicator/SP.DYN.LE00.IN?format=json&date=2021&per_page=20000'
### 3: only usa 2000-2021
USA2020_2021_gdp_endpoint = 'https://api.worldbank.org/v2/country/USA/indicator/NY.GDP.PCAP.PP.KD?format=json&date=2000:2021&per_page=20000'
USA2020_2021_life_endpoint = 'https://api.worldbank.org/v2/country/USA/indicator/SP.DYN.LE00.IN?format=json&date=2000:2021&per_page=20000'
### 4: only usa:can 2000-2021
USACAN2020_2021_gdp_endpoint = 'https://api.worldbank.org/v2/country/usa;can/indicator/NY.GDP.PCAP.PP.KD?format=json&date=2000:2021&per_page=2000'
USACAN2020_2021_life_endpoint = 'https://api.worldbank.org/v2/country/usa;can/indicator/SP.DYN.LE00.IN?format=json&date=2000:2021&per_page=20000'


indicators = 'https://api.worldbank.org/v2/indicator?format=json&per_page=30000'
```

# all country

```python
### what we wanna look up - GDP per capita, PPP (constant 2017 international $)
response = requests.get(gdp_endpoint)
r = response.json()

### make df - the first row is metadata not intended for df
data = r[1]
gdp_df = pd.json_normalize(data)

gdp_df
```

```python
### what we wanna look up - Life expectancy at birth, total (years)
response = requests.get(life_endpoint)
r = response.json()

### make df - the first row is metadata not intended for df
data = r[1]
life_df = pd.json_normalize(data)

### drops
life_df = life_df.dropna(subset=['value'])

life_df
```

```python
### format our life_df for merging
life_df.columns = life_df.columns.str.replace('.', '_')

life_df = life_df[[
    'country_value',
    'countryiso3code',
    'date',
    'value'
]]

life_df = life_df.rename(columns={
    'value': 'life_expectancy',
    'country_value': 'country'
})

### format our gdp_df for merging
gdp_df.columns = gdp_df.columns.str.replace('.', '_')

gdp_df = gdp_df.dropna(subset=['value'])

gdp_df = gdp_df[[
    'country_value',
    'countryiso3code',
    'country_id',
    'date',
    'value'
]]

gdp_df = gdp_df.rename(columns={
    'value': 'gdp',
    'country_value': 'country'
})
```

```python
### merge and format the 2 df into gdp_life_df
gdp_life_df = life_df.merge(gdp_df, 'inner', ['countryiso3code', 'date'])
gdp_life_df = gdp_life_df.rename(columns={"country_x": "country"}).drop(columns=["country_y"])
gdp_life_df = gdp_life_df[['country', 'countryiso3code', 'country_id', 'date', 'gdp', 'life_expectancy']]
gdp_life_df
```

```python
### what we wanna look up - all country info
response = requests.get(country_info_endpoint)
r = response.json()

### make df - the first row is metadata not intended for df
data = r[1]
country_df = pd.json_normalize(data)

### looks like any row that doesnt have lat or lon is an aggregate/non-country
country_df = country_df[country_df['longitude'] != '']

country_df = country_df.rename(columns={"name": "country", "capitalCity": "capital_city", "iso2Code":"iso2code"})
country_df
```

```python
### merge and format that thang
gdp_life_country_df = gdp_life_df.merge(country_df, 'inner', 'country')

gdp_life_country_df = gdp_life_country_df.rename(columns={"countryiso3code":"iso3code"}).drop(columns=['id', 'country_id'])
gdp_life_country_df = gdp_life_country_df[[
    "country",
    "date",
    "gdp",
    "life_expectancy",
    "capital_city",
    "incomeLevel.value",
    "adminregion.value",
    "iso2code",
    "iso3code",
    
]]
gdp_life_country_df
```
